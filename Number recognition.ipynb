```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt


class ImgObj:
    def __init__(self, img):
        self.img = img
        #         self.small_img = None
        self.matched_num = None
        self.shape = {"shape": None, "shape_cnt": 0, "shape_size": 0, "vertices": []}
        self.corner = {"corners": 0}


# 사진을 정규화 시킵니다.
# return: 정규화된 사진
def regularize_image(img, contours):
    # find a bounding box
    cnt = contours[0]
    x, y, w, h = cv.boundingRect(cnt)

    # crop the image
    img_reg = img.copy()
    img_reg = img_reg[y:y + h, x: x + w]

    # scale the image
    height, width, _ = img_reg.shape
    height = int((height * 30) / width)
    width = 30
    img_reg = cv.resize(img_reg, dsize=(width, height), interpolation=cv.INTER_AREA)
    img_reg2 = cv.copyMakeBorder(img_reg, 5, 5, 5, 5, cv.BORDER_CONSTANT, value=[255, 255, 255])
    #     cv.imshow("reg", img_reg)
    #     cv.waitKey(0)
    return img_reg2


# corner에 점을 찍습니다.
# return: corner좌표
def detect_corner(img, howmany, percent, euclidean):
    imgray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    imgray = cv.bitwise_not(imgray)
    #     cv.imshow("img2", imgray)
    #     cv.waitKey(0)
    # blur
    #     imblur = cv.blur(imgray, (2,2))
    # calulate corners
    corners = cv.goodFeaturesToTrack(imgray, howmany, percent, euclidean)
    corners = np.int0(corners)
    position = []
    h, w, _ = img.shape
    #     print(w)
    img_re = img[5: h - 5, 5: w - 5]
    #print(img_re.shape)
    #     cv.imshow("why", img)
    #     cv.waitKey(0)
    # draw the coners
    f1 = plt.figure()
    for i in corners:
        x, y = i.ravel()
        position.append((x - 5, y - 5))
    #         cv.circle(img_re,(x-5,y-5),2,255,-1)
    graph1 = plt.subplot(2, 2, 1)
    plt.imshow(img_re)
    plt.show()
    #     print(corners)
    #print(corners)
    return position

def detect_line(img):
    height, _, _ = img.shape
    im_new = img.copy()
    imgray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    imgray = cv.bitwise_not(imgray)
    edges = cv.Canny(imgray, 50, 150, apertureSize=3)
    
    lines = cv.HoughLinesP(edges, 1, np.pi, 30, int(height/2), 10)
#     print(lines)
    if np.any(lines != None):
        lines = lines
#         for x1, y1, x2, y2 in lines[0]:
#             cv.line(im_new, (x1, y1), (x2, y2), (0,255,0), 2)
    else:
        lines = []
#     plt.imshow(im_new)
    return lines

# 도형이 있는지 찾고, 도형이름, 도형갯수, 꼭짓점좌표를 찾습니다.
# return: 도형이름, 도형갯수, 꼭짓점 갯수, 꼭짓점 좌표
def detect_shape(img):
    # regularize and contour again
    shape = None
    shape_cnt = 0
    shape_size = 0
    img = cv.bitwise_not(img)
    imgray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    kernel = np.ones((2,2), np.uint8)
    imdilate = cv.dilate(imgray, kernel, iterations=1)
    imblur = cv.blur(imdilate, (2,2))
    ret, imbinary = cv.threshold(imblur, 127, 255, 0)
    contours, hierarchy = cv.findContours(imbinary, cv.RETR_CCOMP, cv.CHAIN_APPROX_SIMPLE)
    # 안쪽 선 contour 만 씁니다.
    vertices = []
    for b, cnt in enumerate(contours):
        if hierarchy[0, b, 3] != -1:  # if it has a parent
            epsilon = 0.07 * cv.arcLength(cnt, True)
            approx = cv.approxPolyDP(cnt, epsilon, True)
            size = len(approx)

            vertices = [tuple(approx[0][0]), tuple(approx[2][0])]
            #             cv.line(img, tuple(approx[0][0]), tuple(approx[1][0]), (0, 255, 0), 3)
            #             cv.line(img, tuple(approx[0][0]), tuple(approx[1][0]), (0, 255, 0), 3)
            #             cv.line(img, tuple(approx[1][0]), tuple(approx[2][0]), (0, 255, 0), 3)
            #             print(tuple(approx[1][0]))
            #             cv.line(img, tuple(approx[0][0]), tuple(approx[size-1][0]), (0, 255, 0), 3)
            #             for k in range(size-1):
            #                 print(tuple(approx[k][0]), tuple(approx[k+1][0]))
            #                 cv.line(img, tuple(approx[k][0]), tuple(approx[k+1][0]), (0, 255, 0), 3)

            max_ver = vertices[1][1]
            for k in range(3, size - 1):
                if approx[k][0][1] > max_ver:
                    vertices[1] = tuple(approx[k][0])
                    max_ver = approx[k][0][1]
                    print(approx[k][0][1], k)
            #             cv.line(img, vertices[0], vertices[1], (0, 255, 0), 3)

            if cv.isContourConvex(approx):
                shape_size = size
                shape_cnt += 1
                if size == 3:
                    shape = "tri"
                elif size == 4:
                    shape = "rect"
                else:
                    shape = "else"
            else:
                shape = None
    return shape, shape_cnt, shape_size, vertices



# 사진 object 의 속성을 받습니다.
# shape정하고, corner점 정하기.
def get_attr(filename):
    src = cv.imread(filename, cv.IMREAD_COLOR)
    img_obj = ImgObj(src)
    dst = img_obj.img.copy()

    # image processing
    imgray = cv.cvtColor(dst, cv.COLOR_BGR2GRAY)
    ret, imbinary = cv.threshold(imgray, 127, 255, cv.THRESH_BINARY_INV | cv.THRESH_OTSU)
    contours, hierarchy = cv.findContours(imbinary, cv.RETR_CCOMP, cv.CHAIN_APPROX_SIMPLE)

    imreg = regularize_image(dst, contours)
    img_obj.img = imreg
    img_obj.shape["shape"], img_obj.shape["shape_cnt"], img_obj.shape["shape_size"], img_obj.shape[
        "vertices"] = detect_shape(imreg)
    img_obj.corner["corners"] = detect_corner(imreg, 3, 0.1, 15)

    return img_obj


# obj의 속성들을 가지고, 어떤 숫자인지 판단합니다.
def detect_number(filename):
    img_obj = get_attr(filename)
    # if it has a shape
    if img_obj.shape["shape"] != None:
        # if it has two shapes -> 8
        if img_obj.shape["shape_cnt"] == 2:
            img_obj.matched_num = "8"
        else:
            # if it has a triangle -> 4
            if img_obj.shape["shape"] == "tri":
                img_obj.matched_num = "4"
            else:
                vert = img_obj.shape["vertices"]
                verts_mean = (vert[0][1] + vert[1][1]) / 2
                print(vert[0][1], " ", vert[1][1])
                img_width, _, _ = img_obj.img.shape
                # the mean of highest vertex and lowest verext
                # the mean is close to the middle --> 0
                if int(img_width / 5 * 2) <= verts_mean <= int(img_width / 5 * 3):
                    img_obj.matched_num = "0"
                # the mean is below to the middle --> 9
                elif verts_mean < int(img_width / 5 * 2):
                    img_obj.matched_num = "9"
                # the mean is above to the middle
                elif verts_mean > int(img_width / 5 * 3):
                    img_obj.matched_num = "6"
    else:
        height, width, _ = img_obj.img.shape
        height -= 10
        width -= 10
        h_mid = round(height * 0.35)
        l_mid = round(height * 0.75)
        first_cnt = 0
        third_cnt_a = 0
        third_cnt_b = 0
        forth_cnt = 0
        l_cnt = 0

        position = img_obj.corner["corners"]
        print("정렬 전 positon: ",position)
        position = sorted(position, key=lambda x: x[1])
        print("position:", position)

        feature_2 = [(pos[0], pos[1]) for pos in position if pos[0] < int(width / 5) and pos[1] > int(height * 4 / 5 +4)]
        feature_5 = [(pos[0], pos[1]) for pos in position if pos[0] > int(width / 2)]
        #print(feature_5)
        #         filter(lambda x: (x[0]<int(width/4) and x[1]>int(height*3/4)), position)
        #print("feature2=", feature_2)


        pos_y = [pos[1] for pos in position]
        #         print("y=", pos_y)

        #pos_y = sorted(pos_y)
        print("sorted pos_y=", pos_y)
        if (int(h_mid) <= pos_y[1] <= int(l_mid) or int(h_mid) <= pos_y[len(pos_y) - 1] <= int(l_mid)) and len(feature_2) == 0:
            if ((pos_y[0] <= int(height * 0.40)) and (pos_y[1] <= int(height * 0.40))) and len(feature_5) <= 1:
                img_obj.matched_num = "5"
                #print("matched number!!!! ====>>>>  " + img_obj.matched_num + "  <<<<<=====")
            else:
                img_obj.matched_num = "3"
                
                #print("matched number!!!! ====>>>>  " + img_obj.matched_num + "  <<<<<=====")

        else:
            for y in pos_y:
                if y >= 30:
                    third_cnt_a = third_cnt_a + 1
                if y >= 10 and y < 30:
                    third_cnt_b = third_cnt_b + 1
            if position[len(position) - 1][0] < 15:
                forth_cnt = forth_cnt + 1

            if first_cnt == 0:
                if third_cnt_a == 2 and third_cnt_b == 1:
                    img_obj.matched_num = "2"
                else:
                    if len(detect_line(img_obj.img)) != 0:
                        img_obj.matched_num = "1"
#                         if forth_cnt == 0 or third_cnt_a == 2:
                        
                    else:
                        img_obj.matched_num = "7"
    return img_obj.matched_num
    
    
    print("matched number!!!! ====>>>>  " + img_obj.matched_num + "  <<<<<=====")
    


# 함수들을 부릅니다.
cnt = 0 
failed = []
for i in range(0,12):
    for j in range(10):
            test_num = str(i) + str(j)
            filename = "test" + test_num + ".png"
            print("***filename***", filename)
            matched_num= detect_number(filename)
            if str(j) != matched_num:
                cnt = cnt + 1
                failed.append([filename, matched_num])
print("정답률 = ", (1-cnt/120.0)*100)
print("틀린것 = ")
for f in failed:
    print("       ", f[0], " --> ", f[1])
    
    
```
